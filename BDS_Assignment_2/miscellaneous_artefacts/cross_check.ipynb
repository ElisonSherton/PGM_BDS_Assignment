```python
df["InvoiceDate"] = pd.to_datetime(df.InvoiceDate)
```

```python
def week_year_format(row):
    return f"Year {row.InvoiceDate.year}; Week {row.InvoiceDate.week:0>2}"

def month_year_format(row):
    return f"Year {row.InvoiceDate.year}; Month {row.InvoiceDate.month:0>2}"
```

```python
df["Week"] = df.apply(lambda x: week_year_format(x), axis = 1)
df["Month"] = df.apply(lambda x: month_year_format(x), axis = 1)
df["InvoiceDate"] = df["InvoiceDate"].dt.strftime("%d/%m/%Y")
```

```python
df.head(5)
```

```python
# Weekwise total, average and number of transactions
daywise_summary = []
for k, g in df.groupby(by = "InvoiceDate", sort = False):
    total_transactions = len(g)
    total_transaction_amount = g.Price.sum()
    average_transaction_amount = g.Price.mean()
    daywise_summary.append([k, total_transaction_amount, total_transactions, average_transaction_amount])
df_daywise = pd.DataFrame(daywise_summary, columns = ["Day", "Total Transaction Value", "# Transactions", 
                                                        "Average Transaction Value"])
df_daywise.head(10)
```

```python
# Weekwise total, average and number of transactions
weekwise_summary = []
for k, g in df.groupby(by = "Week"):
    total_transactions = len(g)
    total_transaction_amount = g.Price.sum()
    average_transaction_amount = g.Price.mean()
    weekwise_summary.append([k, total_transaction_amount, total_transactions, average_transaction_amount])
df_weekwise = pd.DataFrame(weekwise_summary, columns = ["Week", "Total Transaction Value", "# Transactions", 
                                                        "Average Transaction Value"])
df_weekwise.head(10)
```

```python
# Weekwise total, average and number of transactions
monthwise_summary = []
for k, g in df.groupby(by = "Month"):
    total_transactions = len(g)
    total_transaction_amount = g.Price.sum()
    average_transaction_amount = g.Price.mean()
    monthwise_summary.append([k, total_transaction_amount, total_transactions, average_transaction_amount])
df_monthwise = pd.DataFrame(monthwise_summary, columns = ["Month", "Total Transaction Value", "# Transactions", 
                                                        "Average Transaction Value"])
df_monthwise
```
