```python
from collections import Iterable
def flatten(items, ignore_types=(str, bytes)):
    for x in items:
        if isinstance(x, Iterable) and not isinstance(x, ignore_types):
            yield from flatten(x)
        else:
            yield x
items = [1, 2, [3, 4, [5, 6], 7], 8]
# Produces 1 2 3 4 5 6 7 8
```

```python
import random, math
i = 0
a = int(100 * random.random())
while i <= 10000:
    a = math.sqrt(2 * a + 24)
    print(a)
    i += 1
```

```python
# Code to execute in an independent thread
from threading import Thread
from datetime import datetime
import time

def compute_mean(processing_time):
    time.sleep(processing_time)
    print(f"Done mean computation at {datetime.now()}")

def compute_standard_deviation(processing_time):
    time.sleep(processing_time)
    print(f"Done median computation at {datetime.now()}")

t_mean = Thread(target = compute_mean, args = (5, ))
t_stdv = Thread(target = compute_standard_deviation, args = (5, ))

t_mean.start(); t_stdv.start()
```

```python
from threading import Thread
from datetime import datetime
import time

def compute_mean(processing_time):
    time.sleep(processing_time)
    print(f"Done mean computation at   {datetime.now()}")
    return 1

def compute_standard_deviation(processing_time):
    time.sleep(processing_time)
    print(f"Done median computation at {datetime.now()}")
    return -1

t_mean = Thread(target = compute_mean, args = (5, ))
t_stdv = Thread(target = compute_standard_deviation, args = (5, ))

t_mean.start(); t_stdv.start(); t_stdv.join(); t_mean.join();
```
